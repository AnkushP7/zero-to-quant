```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
monthly = df.resample("M").agg({"Close": "mean", "Volume": "sum"})
print(monthly)
```

```python
df['year'] = df.index.year
df['month'] = df.index.month
avg_yearly_close = df.groupby('year').Close.mean().sort_values()
print(avg_yearly_close)
```

```python
vol_year_month = df.groupby(['year','month']).Volume.sum()
print(vol_year_month)
```

```python
close_yearly_monthly_data = df.groupby(['year', 'month']).Close.agg([len,'std','mean'])
print(close_yearly_monthly_data)
```
